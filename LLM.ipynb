#Import Libraries

```python
import pandas as pd
import numpy as np
import faiss
from sentence_transformers import SentenceTransformer
```

#code:

```python
# Load the pretrained PatentBERT model (SBERT variant)
model = SentenceTransformer("patent/sbert-all-MiniLM-L6-v2")
```

```python
# Load the dataset
df = pd.read_excel(r"C:\Users\israh\Desktop\Masters\THESIS\code\Patent_database\patent100.xlsx")
print(df.head())
```

```python
# Combine title and abstract for better embedding representation
#df["text"] = df["patent_id"] + ". " + df["patent_abstract"]

# Generate embeddings for the patent texts
embeddings = model.encode(df["patent_abstract"].tolist(), convert_to_numpy=True)
```

```python
# Initialize FAISS index for fast similarity search
dim = embeddings.shape[1]
index = faiss.IndexFlatL2(dim)
index.add(embeddings)
```

```python
def search_patents(query, top_k=5):
    """Search for similar patents based on a query."""
    query_embedding = model.encode([query], convert_to_numpy=True)
    distances, indices = index.search(query_embedding, top_k)
    results = df.iloc[indices[0]]
    return results[["patent_id", "patent_abstract"]]
```

```python
# query here and results
if __name__ == "__main__":
    query = "laser detecting and rainging system"
    results = search_patents(query)
    print(results)
```
