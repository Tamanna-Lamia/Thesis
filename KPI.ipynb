#Import Libraries

```python
import torch
from codecarbon import EmissionsTracker
from transformers import AutoModelForSequenceClassification, AutoModelForSeq2SeqLM, AutoTokenizer, pipeline,BertTokenizerFast,AutoModelForMaskedLM
```

#Code Body:

```python
tracker = EmissionsTracker(project_name="LLM_KPI", measure_power_secs=10)
```

```python
sample_text = "A new method for semiconductor manufacturing involving nanostructures."
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
#test if LLM model is accessible
model_name = "dheerajpai/patentbert"

try:
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    print(f"✅ Model {model_name} is public and accessible!")
except Exception as e:
    print(f" Error loading {model_name}: {e}")
```

# PatentBERT

```python
tracker.start_task("PatentBERT")

# Load Model and Tokenizer
tokenizer = AutoTokenizer.from_pretrained("dheerajpai/patentbert")
model = AutoModelForMaskedLM.from_pretrained("dheerajpai/patentbert")

# Tokenize Input and Run Inference
inputs = tokenizer(sample_text, return_tensors="pt").to(device)
with torch.no_grad():
    _ = model(**inputs)

# Stop Tracker and Print Results
emissions_patentbert = tracker.stop_task()
#print(f"PatentBERT CO₂ Emissions: {emissions_patentbert.get('emissions', 0):.6f} kg CO₂")

```

#SciBERT

```python
tracker.start_task("SciBERT")

# Load Model and Tokenizer
tokenizer = AutoTokenizer.from_pretrained("allenai/scibert_scivocab_uncased")
model = AutoModelForSequenceClassification.from_pretrained("allenai/scibert_scivocab_uncased").to(device)

# Tokenize Input and Run Inference
inputs = tokenizer(sample_text, return_tensors="pt").to(device)
with torch.no_grad():
    _ = model(**inputs)

# Stop Tracker and Print Results
emissions_scibert = tracker.stop_task()
#print(f"SciBERT CO₂ Emissions: {emissions_scibert:.6f} kg CO₂")
```

#Flan T5

```python
tracker.start_task("FlanT5")

# Load Model and Tokenizer
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small").to(device)

# Tokenize Input and Run Inference
inputs = tokenizer(sample_text, return_tensors="pt").to(device)
with torch.no_grad():
    _ = model.generate(**inputs)

# Stop Tracker and Print Results
emissions_flan_t5 = tracker.stop_task()
#print(f"Flan-T5 CO₂ Emissions: {emissions_flan_t5:.6f} kg CO₂")
```

#Falcon

```python
tracker.start_task("Falcon")

# Load Model and Tokenizer
tokenizer = AutoTokenizer.from_pretrained("tiiuae/falcon-7b")
model = AutoModelForSequenceClassification.from_pretrained("tiiuae/falcon-7b").to(device)

# Tokenize Input and Run Inference
inputs = tokenizer(sample_text, return_tensors="pt").to(device)
with torch.no_grad():
    _ = model(**inputs)

# Stop Tracker and Print Results
emissions_falcon = tracker.stop_task()
#print(f"Falcon CO₂ Emissions: {emissions_falcon:.6f} kg CO₂")
```
